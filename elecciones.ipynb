```python
def cargarInformacion(nombreArchivo):
    """crea un diccionario donde las claves corresponden a un numero secuencial y los valores un diccionario de preguntas con sus respuestas"""
    archivo = open(nombreArchivo)
    d = {i+1 : {'pregunta'+str(j+1): respuesta for j,respuesta in enumerate(linea.rstrip().split(";"))} for i,linea in enumerate(archivo)}
    archivo.close()
    return d

def obtenerMedioMasPopular(d):
    """Cuenta los medios y devuelve el más popular"""
    MasPopular = max({d[i]['pregunta7']:len([j['pregunta7'] for j in list(d.values()) if j['pregunta7'] == d[i]['pregunta7']]) for i in d}.items(), key=lambda x:x[1])
    return 'El medio más popular ha sido %s con %s ocurrencias' % MasPopular

def candidatoPreferidoXRangoDeEdad(d):
    """Crea una tabla que indica las preferencias de los votantes por edades"""
    
    dict = {d[i]['pregunta2']:{d[j]['pregunta5']:len([k['pregunta5'] for k in list(d.values()) if k['pregunta2'] == d[i]['pregunta2'] and k['pregunta5'] == d[j]['pregunta5']]) for j in d} for i in d}
    
    print(59*"=")
    print('{0:15s} {1:35} {2:>5}'.format('Edades', 'Candidatos', 'Votos'))
    print(59*"=")
    for i in sorted(dict):
        maximo =  max(dict[i].items(), key=lambda x:x[1])[1]
        nombres = ''
        for j in dict[i].items():
            if j[1] == maximo:
                if len(nombres) == 0:
                    nombres = j[0]
                else:
                    nombres += ', '+j[0]
        print('{0:15s} {1:35} {2:>5}'.format(i, nombres, maximo))
    print(59*"=")
    return str()

def candidatoPreferidoXRangoDeEdad2(datos):
    """Crea una tabla que indica las preferencias de los votantes por edades"""
    from numpy import unique
    print(57*"=")
    print('{0:15} {1:35} {2:>5}'.format('Edades', 'Candidatos', 'Votos'))
    print(57*"=")

    for edades in sorted({datos[i]['pregunta2'] for i in datos}):
        l = [datos[i]['pregunta5'] for i in datos if datos[i]['pregunta2']== edades]
        candidatos, votos = unique(l, return_counts=True)

        nombres = ''
        for i,j in enumerate(votos):
            if j == max(votos):
                if len(nombres) == 0:
                    nombres = candidatos[i]
                else:
                    nombres += ', '+candidatos[i]
        print('{0:15} {1:35} {2:>5}'.format(edades, nombres, max(votos)))
    print(57*"=")
```
